# Content Based Recommendation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
%matplotlib inline
```

```python
movies_df=pd.read_csv('./datasets/moviedataset/movies.csv')
ratings_df=pd.read_csv('./datasets/moviedataset/ratings.csv')
movies_df.head(5)
```

```python
ratings_df.head(5)
```

### Extracting Year from title

```python
movies_df['year']=movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
movies_df['year']=movies_df.year.str.extract('(\d\d\d\d)',expand=False) # removing paranthesis
movies_df['title']=movies_df.title.str.replace('(\(\d\d\d\d\))','') # removing year from title
movies_df['title']=movies_df['title'].apply(lambda x: x.strip())
movies_df.head(5)
```

### Splitting geners

```python
movies_df['genres']=movies_df.genres.str.split('|')
movies_df.head(5)
```

### Dataset of Movies with ratings(1) for each genres

```python
moviesWithGenre_df=movies_df.copy()

for index,row in movies_df.iterrows():
    for genre in row['genres']:
        moviesWithGenre_df.at[index,genre]=1

moviesWithGenre_df = moviesWithGenre_df.fillna(0)
moviesWithGenre_df.head(5)
```

### Ratings

```python
ratings_df = ratings_df.drop('timestamp',1)
ratings_df.head(5)
```

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
# adding movieId to list
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
inputMovies= pd.merge(inputId,inputMovies)
inputMovies=inputMovies.drop('genres',1).drop('year',1)
inputMovies
```

```python
# selecting input movies from data
userMovies = moviesWithGenre_df[moviesWithGenre_df['movieId'].isin(inputMovies['movieId'].tolist())]
userMovies
```

```python
# dropping title, genre, year, ID
userMovies=userMovies.reset_index(drop=True)
userGenreTable=userMovies.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
userGenreTable
```

```python
inputMovies['rating']
```

```python
userProfile=userGenreTable.transpose().dot(inputMovies['rating'])
userProfile
```

```python
#remaining movies other than the rated ones...
genreTable=moviesWithGenre_df.set_index(moviesWithGenre_df['movieId'])
genreTable = genreTable.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
genreTable
```

```python
recommendation_df=((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendation_df
```

```python
recommendation_df = recommendation_df.sort_values(ascending=False)
recommendation_df
# percentage of recommendation
```

```python
recommendation_df2=genreTable.dot(userProfile)
recommendation_df2.sort_values(ascending=False) 
#gives the recommendation socre out of (Sum of user ratings based on genre)
```

### Selecting the movies based on Recommendations
